# Lab Task 1:

Select only two categories:

### Category 1: Healthy

Use 120 images for training, 30 images for validation, and 12 images for testing.

### Category 2: Unhealthy (combine the other three classes)

From each of the three remaining classes, select:

40 images for training,

10 images for validation,

4 images for testing.


```python
import os, shutil

original_dataset_dir = r'E:\ELECTRICAL ENGINEERIGH(COMNICATN)\Semester-8th\ML Lab\lab11\archive\plantvillage dataset\color'
base_dir = r'E:\ELECTRICAL ENGINEERIGH(COMNICATN)\Semester-8th\ML Lab\lab11\dataset_task1'
os.makedirs(base_dir, exist_ok=True)

# Subdirectories
for split in ['train', 'validation', 'test']:
    for category in ['healthy', 'unhealthy']:
        os.makedirs(os.path.join(base_dir, split, category), exist_ok=True)

# --- HEALTHY ---
src_healthy = os.path.join(original_dataset_dir, 'Corn_(maize)___healthy')
files = os.listdir(src_healthy)

# TRAIN: 120
for i in range(120):
    shutil.copyfile(os.path.join(src_healthy, files[i]), os.path.join(base_dir, 'train/healthy', f'healthy.{i}.jpg'))

# VALIDATION: 30
for i in range(120, 150):
    shutil.copyfile(os.path.join(src_healthy, files[i]), os.path.join(base_dir, 'validation/healthy', f'healthy.{i}.jpg'))

# TEST: 12
for i in range(150, 162):
    shutil.copyfile(os.path.join(src_healthy, files[i]), os.path.join(base_dir, 'test/healthy', f'healthy.{i}.jpg'))

# --- UNHEALTHY from 3 Classes ---
classes = {
    'Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot': 'cercospora',
    'Corn_(maize)___Common_rust_': 'common_rust',
    'Corn_(maize)___Northern_Leaf_Blight': 'northern_leaf_blight'
}

for cls_folder, prefix in classes.items():
    src = os.path.join(original_dataset_dir, cls_folder)
    files = os.listdir(src)

    # TRAIN: 40 per class
    for i in range(40):
        shutil.copyfile(os.path.join(src, files[i]),
                        os.path.join(base_dir, 'train/unhealthy', f'{prefix}.{i}.jpg'))

    # VALIDATION: 10 per class
    for i in range(40, 50):
        shutil.copyfile(os.path.join(src, files[i]),
                        os.path.join(base_dir, 'validation/unhealthy', f'{prefix}.{i}.jpg'))

    # TEST: 4 per class
    for i in range(50, 54):
        shutil.copyfile(os.path.join(src, files[i]),
                        os.path.join(base_dir, 'test/unhealthy', f'{prefix}.{i}.jpg'))
```
