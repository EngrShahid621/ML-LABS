```python
import pandas as pd
import numpy as np

# Sample loading of your dataset (replace this with your actual loading method)
# cleaned_citylearn_data = pd.read_csv('your_file.csv')

# For demonstration, assuming cleaned_citylearn_data is already loaded as DataFrame

def detect_outliers_iqr(data):
    """
    Detect outliers in the specified column using the IQR method.
    Replace outliers with NaN in the original DataFrame.

    Args:
        data (pd.DataFrame): Input DataFrame.
        column_name (str): Column to detect outliers.

    Returns:
        outliers (list): List of detected outlier values.
        data (pd.DataFrame): DataFrame with outliers replaced by NaN.
        idx_list (list): List of indices where outliers were found.
    """
    q1 = data['non_shiftable_load'].quantile(0.25)
    q3 = data['non_shiftable_load'].quantile(0.75)
    IQR = q3 - q1
    lwr_bound = q1 - 1.5 * IQR
    upr_bound = q3 + 1.5 * IQR

    outliers = []
    idx_list = []

    for idx, val in data['non_shiftable_load'].items():
        if val < lwr_bound or val > upr_bound:
            outliers.append(val)
            idx_list.append(idx)

    # Replace outliers with NaN
    data.loc[data['non_shiftable_load'] < lwr_bound, 'non_shiftable_load'] = np.nan
    data.loc[data['non_shiftable_load'] > upr_bound, 'non_shiftable_load'] = np.nan

    return outliers, data, idx_list

# --- Main Code ---

# Detect outliers in 'non_shiftable_load'
outliers, cleaned_citylearn_data_outlr, outlier_indices = detect_outliers_iqr(data)

print(f"Outliers detected in 'non_shiftable_load': {outliers}")
print(f"Number of outliers: {len(outliers)}")

# Check missing values after outlier replacement
missing_count = cleaned_citylearn_data_outlr['non_shiftable_load'].isnull().sum()
print(f"Missing values after replacing outliers: {missing_count}")

# Interpolate missing values in 'non_shiftable_load'
# Since you may not have a datetime index, use method='linear' or 'pad' interpolation
# If you have datetime index, you can use method='time' interpolation

cleaned_citylearn_data_outlr['non_shiftable_load'].interpolate(method='linear', inplace=True)

# Verify if missing values remain after interpolation
missing_after_interp =cleaned_citylearn_data_outlr['non_shiftable_load'].isnull().sum()
print(f"Missing values after interpolation: {missing_after_interp}")

# Optionally, you can forward fill or backward fill remaining NaNs (if any)
if missing_after_interp > 0:
    cleaned_citylearn_data_outlr['non_shiftable_load'].fillna(method='ffill', inplace=True)
    cleaned_citylearn_data_outlr['non_shiftable_load'].fillna(method='bfill', inplace=True)
    print(f"Missing values after forward/backward fill: {cleaned_citylearn_data_outlr['non_shiftable_load'].isnull().sum()}")

# Save the cleaned dataset to CSV
cleaned_citylearn_data_outlr.to_csv(r'C:\Users\PMLS\ml\lab4\cleaned_citylearn_data_no_outliers.csv', index=False)
print("Cleaned dataset saved successfully.")
```

```python
import pandas as pd
import numpy as np

# Load your dataset
data1 = pd.read_csv('citylearn_data_outlr_reordered.csv')  # change filename if needed

# Columns to exclude
exclude_cols = ['non_shiftable_load', 'hour', 'month', 'day_type']

# Select numeric columns for outlier detection (excluding specified ones)
columns_to_check = [col for col in data1.columns if col not in exclude_cols and pd.api.types.is_numeric_dtype(data1[col])]

print(f"Columns for outlier detection: {columns_to_check}")

def detect_and_handle_outliers_iqr(data1, column):
    """
    Detects and replaces outliers in a specific column using the IQR method.
    Replaces outliers with NaN and interpolates missing values.
    """
    q1 = data1[column].quantile(0.25)
    q3 = data1[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Detect outliers
    outliers = data1[(data1[column] < lower_bound) | (data1[column] > upper_bound)]
    print(f"{column}: Found {outliers.shape[0]} outliers.")

    # Replace with NaN
    df.loc[data1[column] < lower_bound, column] = np.nan
    df.loc[data1[column] > upper_bound, column] = np.nan

    # Interpolate missing values linearly
    data1[column].interpolate(method='linear', inplace=True)

    # If still NaN exists, use forward/backward fill
    if data1[column].isnull().sum() > 0:
        data1[column].fillna(method='ffill', inplace=True)
        data1[column].fillna(method='bfill', inplace=True)

    return data1

# Process each column
for col in columns_to_check:
    data2 = detect_and_handle_outliers_iqr(data1, col)

# Save cleaned dataset
data2.to_csv(r'C:\Users\PMLS\shahid (ML-lab)\cleaned_citylearn_data_all_outliers_handled.csv', index=False)
print("Cleaned dataset saved successfully.")
```

```python
from scipy import stats

# Function to cap outliers
def cap_outliers(df, column, threshold=3):
    z = np.abs(stats.zscore(df[column]))
    df[column] = np.where(z > threshold, df[column].median(), df[column])
    return df

# Apply to numerical columns (excluding target and categorical)
numerical_cols = processed_df.select_dtypes(include=[np.number]).columns.tolist()
numerical_cols.remove('non_shiftable_load')  # Don't modify target

for col in numerical_cols:
    processed_df = cap_outliers(processed_df, col)
```

```python
# Cyclical encoding for temporal features
processed_df['hour_sin'] = np.sin(2 * np.pi * processed_df['hour']/24)
processed_df['hour_cos'] = np.cos(2 * np.pi * processed_df['hour']/24)
processed_df['month_sin'] = np.sin(2 * np.pi * processed_df['month']/12)
processed_df['month_cos'] = np.cos(2 * np.pi * processed_df['month']/12)

# Interaction terms
processed_df['temp_load_interaction'] = processed_df['outdoor_dry_bulb_temperature'] * processed_df['cooling_demand']
processed_df['solar_price_interaction'] = processed_df['solar_generation'] * processed_df['electricity_pricing']

# Time-based aggregations (example)
processed_df['rolling_avg_load_24h'] = processed_df['non_shiftable_load'].rolling(24).mean().shift(1)
```
